# Find a fake news and count (amazon.csv)

```python
amazon=pd.read_csv("D:\\csv\\amazon.csv")

positive=amazon["Positive"].tolist()

fig=plt.figure(figsize=(7, 4))
ax=fig.add_subplot()

non_fake=0
fake=0

for i in positive:
    if i==1:
        non_fake=non_fake+1
    if i==0:
        fake=fake+1
        
dic=[non_fake]
label=['Non-Fake']

ax.bar(label, dic, label=f'Non-Fake - {non_fake}')

dic=[fake]
label=['Fake']
ax.bar(label, dic, label=f'Fake - {fake}')


fig1=plt.figure(figsize=(7, 4))
ax1=fig1.add_subplot()

new_graph=[non_fake, fake]
new_graph_label=["Non-Fake", "Fake"]
ax1.pie(new_graph, labels=new_graph_label, autopct='%.2f')

ax.legend()
plt.show()
```

# Output Rating Universities (adm_data.csv)

```python
adm=pd.read_csv("D:\\csv\\adm_data.csv")

rating=adm["University Rating"].tolist()

fig1=plt.figure(figsize=(7,4))
ax1=fig1.add_subplot()

ax1.set_title('Graph of rating universities')
r1=0
r2=0
r3=0
r4=0
r5=0
for i in rating:
    if i==1:
        r1=r1+1
    if i==2:
        r2=r2+1
    if i==3:
        r3=r3+1
    if i==4:
        r4=r4+1
    if i==5:
        r5=r5+1
        
count_ratings=[r1,r2,r3,r4,r5]
count_ratings_label=["Rating 1", "Rating 2", "Rating 3", "Rating 4", "Rating 5"]

ax1.bar(count_ratings_label, count_ratings)

fig=plt.figure(figsize=(7, 4))
ax=fig.add_subplot()
ax.pie(count_ratings, labels=count_ratings_label)
ax.set_title('Pie of rating universities')

plt.show()
```

# Output TOEFL Score University (adm_data.csv)

```python
datas=pd.read_csv("D:\\csv\\adm_data.csv")
toefl_score=datas["TOEFL Score"].tolist()

s=0
s1=0
s2=0
for i in toefl_score:
    if i<100:
        s=s+1
    if i>100 and i<110:
        s1=s1+1
    if i>110:
        s2=s2+1
        
scores=[s,s1,s2]
scores_label=["<100" , ">100 and <110", ">110"]

fig=plt.figure(figsize=(7, 4))
ax=fig.add_subplot()

ax.bar(scores_label, scores)
ax.set_title('TOEFL Score University')

fig1=plt.figure(figsize=(7, 4))
ax1=fig1.add_subplot()
ax1.pie(scores, labels=scores_label, autopct='%.2f')
ax1.set_title('TOEFL Score University')

plt.show()
```

# Output 10-TOP Movies in Netflix

```python
datas=pd.read_csv("D:\\csv\\Best Show by Year Netflix.csv")

movies_ratings=datas["MAIN_GENRE"].tolist()

array_d = {}.fromkeys(movies_ratings, 0)
for a in movies_ratings:
    array_d[a] += 1
count=list(array_d.values())
genre=list(array_d.keys())

fig=plt.figure(figsize=(7,4))
ax=fig.add_subplot()

ax.pie(count, labels=genre, autopct='%.2f')

plt.show()
```

```python
datas=pd.read_csv("D:\\csv\\Best Show by Year Netflix.csv")

movies_ratings=datas["SCORE"].tolist()
movies_title=datas["TITLE"].tolist()

s=dict(zip(movies_title, movies_ratings))

data_sort = dict(sorted(s.items(), key=lambda x:x[1], reverse=True))

movies_sort=list(data_sort.keys())[0:10]
movies_title_sort=list(data_sort.values())[0:10]

print(movies_sort)
print(movies_title_sort)

fig=plt.figure(figsize=(10,4))
ax=fig.add_subplot()

top=[f"#{i}" for i in range(1, 11)]
ax.bar(top, movies_title_sort, width=0.3)

plt.show()
```
